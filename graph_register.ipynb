```python
# 1. all of register draw in one figure 
import matplotlib.pyplot as plt
import os
import numpy as np
import csv
labels = []
#data l
with open('rv5_30_nop+/reg_result_count.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        x = np.arange(34)
        plt.plot(x, line[1:])
        labels.append(line[0])
        plt.xticks(np.arange(0, 29, step=1))

# detail config
plt.legend(labels, 
           loc='best', bbox_to_anchor=[0.5, 1.1],
           columnspacing=1.0, labelspacing=0.0,
           handletextpad=0.0, handlelength=2.0,
           fancybox=True, shadow=True)
plt.grid(True)
plt.yscale('linear')
plt.axvline(17,color='r', linestyle='--', linewidth=3)
plt.text(16, 8, s='reset', fontsize=10,color='r')

#save path
path = 'rv5_30_nop+/figures'
figname = 'all_of_figure.png'
dest = os.path.join(path, figname)
             
#file save 
plt.savefig(dest)
plt.show()
plt.cla()
plt.close()
```

```python
# 2. save each of register in path -> make_plots(path) 
from matplotlib import pyplot as plt, patches
import os
import csv
import numpy as np
def main():
    #path
    make_plots('rv5_30_nop+/figures')
    print('Done.')


def make_plots(basic_path):
    fig, ax = plt.subplots()
    with open('rv5_30_nop+/reg_result_count.csv', newline='') as csvfile:
        reader = csv.reader(csvfile)
        var_count = 0
        for line in reader:
            fig, ax = plt.subplots()
            path = ""
            #range(x): x = cycle number 
            a = range(34)
            plt.xticks(np.arange(0, 34, step=1))
            ax.plot(a, line[1:], label=line[0])
            legend = ax.legend(loc='best', shadow=True, fontsize='x-large')
            
            plt.grid(True)
            #figure size
            plt.rcParams["figure.figsize"] = (14,4)
            plt.axvline(17,color='r', linestyle='--', linewidth=3)
            plt.text(16, 3, s='reset', fontsize=10,color='r')
            plt.axvline(21,color='b', linestyle='--', linewidth=3)
            plt.text(21, 3, s='Add done!', fontsize=10,color='b')
       
            #tendency - incresing (It is not possible decrease) or flat
            tendency = int(line[-1])-int(line[1])
            if tendency==0 :
                path=basic_path+"/flat"
                
            else:
                path=basic_path+"/variation"
                var_count= var_count + 1
            
            
            figname = 'fig_{}.png'.format(line[0])
            dest = os.path.join(path, figname)
            
            
            #file save 
            plt.savefig(dest)
            plt.cla()
            plt.close()
        print("var_count= ", var_count)


main() 
```

```python
#example 1 save 21 figure 
from matplotlib import pyplot as plt, patches
import os


def main():
    for i in range(21):
        print('Batch {}'.format(i))
        make_plots('figures')
    print('Done.')


def make_plots(path):
    fig, ax = plt.subplots()
    for i in range(21):
        x = range(3 * i)
        y = [n * n for n in x]
        ax.add_patch(patches.Rectangle(xy=(i, 1), width=i, height=10))
        plt.step(x, y, linewidth=2, where='mid')
        figname = 'fig_{}.png'.format(i)
        dest = os.path.join(path, figname)
        plt.savefig(dest)  # write image to file
        plt.cla()
    plt.close(fig)


main()
```

```python
# example : line color
import matplotlib.pyplot as plt
import numpy as np

num_plots = 20

# Have a look at the colormaps here and decide which one you'd like:
# http://matplotlib.org/1.2.1/examples/pylab_examples/show_colormaps.html
colormap = plt.cm.gist_ncar
plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.jet(np.linspace(0, 1, num_plots))))

# Plot several different functions...
x = np.arange(10)
labels = []
for i in range(1, num_plots + 1):
    plt.plot(x, i * x + 5 * i)
    labels.append(r'$y = %ix + %i$' % (i, 5*i))

# I'm basically just demonstrating several different legend options here...
plt.legend(labels, ncol=4, loc='upper center', 
           bbox_to_anchor=[0.5, 1.1], 
           columnspacing=1.0, labelspacing=0.0,
           handletextpad=0.0, handlelength=1.5,
           fancybox=True, shadow=True)

plt.show()
```
